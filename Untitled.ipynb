```python
import numpy as np
from matplotlib import pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
all_attributes = []
with open("all_attributes.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        all_attributes.append(line.rstrip())
```

```python
word_embeddings = {}
count = 0
with open("./word_embeddings/glove.6B.300d.txt", "r") as file:
    while True:
        line = file.readline()
        key = line.split(" ")[0]
        val = np.array([float(x) for x in line.split(" ")[1:]])
        word_embeddings[key] = val
        
        count += 1
        if not line:
            print("Total words: {}".format(count))
            break
```

```python
attribute_embeddings = {}
for attr in all_attributes:
    attr_list = attr.split(" ")
    attribute_embeddings[attr] = 0
    for att in attr_list:
        attribute_embeddings[attr] += word_embeddings[att]

labels = list(attribute_embeddings.keys())        
data = np.stack(list(attribute_embeddings.values()))
print(data.shape)
```

```python
attrCategories = []
with open("query_args.txt", "r") as file:
    lines = file.readlines()
    
    for line in lines:
        attrCategories.append(line.rstrip())

cat_embeddings = {}
for cat in attrCategories:
    cat_list = cat.split(" ")
    if cat == "hposition":
        cat_list = ["horizontal", "position"]
    elif cat == "sportActivity":
        cat_list = ["sports", "activity"]
    cat_embeddings[cat] = 0
    for catg in cat_list:
        cat_embeddings[cat] += word_embeddings[catg]
    
cat_labels = list(cat_embeddings.keys())
cat_data = list(cat_embeddings.values())
```

```python
kmeans = KMeans(n_clusters=len(cat_data)-1)
data_cluster = np.argmin(kmeans.fit_transform(data), axis=1)
cluster_means = kmeans.cluster_centers_
```
